la fin.

Exemple d'exécution:

> \# separe.py
>
> L =\["Bonjour" ," " ,"a" ," tout" ," le" ," monde !!" \]   
>
> ...

```
# separe.py
L =["Bonjour" ," " ,"a" ," tout" ," le" ," monde !!" ]
L1 = []
L2 = []
L3 = []
n1 = int(input("n1 = "))
n2 = int(input("n2 = "))
for elt in L :
    if len(elt) < n1 :
        L1 = L1 + [elt]
    elif len(elt) > n2 :
        L3 = L3 + [elt]
    else :
        L2 = L2 + [elt]

print("Chaines de longueur <",n1," : ",L1)
print("Chaines de longueur entre ",n1," et ",n2," : ",L2)
print("Chaines de longueur >",n2," : ",L3)
```

**Exercice 6. Fonctions 1 (3 pts)**

Toutes les fonctions seront écrites dans un fichier **fonctions-1.py**.

1. Écrire une fonction **SomInt(a,b)** qui retourne la somme des nombres
entre $a$ et $b$ (inclus).

2. Écrire  une fonction **U(n)** qui retourne la valeur du $n$ieme terme de
la suite $\left(u_{n}\right)$ définie par $u_{0} = 1$; $u_{n+1} = 3u_{n}+ 1$.

3. Écrire une fonction **NbPaires(ch)** qui retourne le nombre de paires
de caractères consécutifs identiques dans la chaine de caracteres $ch$.

Exemple d'exécution:

> \>\>\> NbPaires("elle est apparue")
>
> 2
>
> \>\>\> NbPaires("creee" )
>
> 2
>
   
</div>

```
def SomInt(a,b):
    S = 0
    for n in range(a,b+1):
        S = S + n
    return S

def U(n):
    u = 1
    if n == 0 :
        return u
    else :
        k = 1
        while k <= n :
            u = 3*u + 1
            k = k + 1
        return u

def NbPaires(ch):
    cpt = 0
    for i in range(len(ch)-1) :
        if ch[i+1] == ch[i] :
            cpt = cpt + 1
    return cpt
```

**Exercice 7. Fonctions 2 (4 pts)**

Toutes les fonctions seront écrites dans un fichier **fonctions-2.py**. 
Un vecteur $\overrightarrow{u} = \left(u_1 , \ldots , u_n\right)$ 
sera representé par une liste de flottants.  
Par exemple, le vecteur $\overrightarrow{u} = \left(1.5, 0, -1\right)$
sera representé par la liste \[1.5,0.0,-1.0\].

1. Écrire une fonction **DifVect(u,v)** qui retourne la différence de deux
vecteurs (de même longueur) $\overrightarrow{u}-\overrightarrow{v} = \left(u_1 - v_1 , \ldots , u_n - v_n \right)$.  
Par exemple **DifVect(\[1,0,3\],\[-1,1,1\])** retournera **\[2,-1,2\]**.

    
2. Écrire une fonction **CarreCoord(u)** qui retourne la somme des
carrées des coordonnees du vecteur $u$.

3. En utilisant les deux fontions précédentes,  
écrire une fonction **Distance(u,v)** qui retourne la distance entre deux vecteurs de même
longueur définie par:

    $\texttt{dist}\left(\overrightarrow{u},\overrightarrow{v}\right) = \sqrt{\left(u_1 - v_1\right)^2+ \cdots + \left(u_n - v_n\right)^2}$

4. Compléter votre script pour que celui-ci affiche le vecteur d'une
liste prédéfinie le plus proche d'un vecteur $u$ prédéfini,  
c'est-à-dire dont la distance à $u$ est la plus petite.
    
Exemple d'exécution:


> \>\>\> # fonction-2.py
>
> l_vect =[[0 ,0 ,0 ,0] , [1,2,-4,3] , [4,7,2,0] , [-1,-2,6,-1], [1,0,-1,3] , [5,-1, -3 ,5]]
>
> u =[1 ,2 ,3 ,4]
>
> ...
>
> Le vecteur le plus proche de [1 , 2, 3, 4] est [1 , 0 , -1 , 3] 
>


   
</div>

```
def DifVect(u,v):
    #assert len(u)==len(v)
    L = []
    for i in range(len(u)) :
        L = L + [u[i]-v[i]]
    return L

def CarreCoord(u):
    S = 0
    for coo in u :
        S = S + coo**2
    return S

def Distance(u,v):
    return (CarreCoord(DifVect(u,v)))**0.5

# fonction-2.py
l_vect =[[0 ,0 ,0 ,0] , [1,2,-4,3] , [4,7,2,0] , [-1,-2,6,-1], [1,0,-1,3] , [5,-1, -3 ,5]]
u =[1 ,2 ,3 ,4]

dmin = Distance(u,l_vect[0])
v = l_vect[0]

for i in range(1,len(l_vect)):
    d = Distance(u,l_vect[i])
    if d < dmin :
        dmin = d
        v = l_vect[i]
        
            
               
print("Le vecteur le plus proche de ",u," est ",v)
```

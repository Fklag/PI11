**Exercice 1. Entrée/Sortie (1 pt)**  

Écrire un programme qui, à partir de la saisie d'un rayon et d'une hauteur, calcule le volume d'un cône droit :  
    
Formule de base : $V = \frac{1}{3} \times \pi \times r^2 \times h$


Exemple d'exécution:

> \>\>\>
>
> Hauteur : 10
>
> Rayon : 20
>
> Volume : 4188.790204786391

</div>

```
from math import *

h=float(input("Hauteur : "))
r=float(input("Rayon : "))
V=1/3 * pi * r**2 * h
print ("Volume : ", V)
```

**Exercice 3. Structures de contrôles (2 pts)**

Écrire un programme qui affiche en fonction d'une valeur saisie l'un des messages suivants :  
- "Ce nombre est pair"  
- "Ce nombre est impair, mais est multiple de 3"
- "Ce nombre n'est ni pair ni multiple de 3"
    
Un nombre **b** est multiple de 3 si le reste de la division de ce nombre par 3 est égal à 0. Ce reste est nommé modulo.   
    
*Syntaxe Python*  
$\%$ permet d'obtenir le modulo en Python



Exemple d'exécution:

> \>\>\>
>
> Nombre : 8
>
> Ce nombre est pair
>
> \>\>\>
>
> Nombre : 9
>
> Ce nombre est impair, mais est multiple de 3
>
> \>\>\>
>
> Nombre : 7
>
> Ce nombre n'est ni pair ni multiple de 3


    
   
</div>

```
nombre = int(input("Nombre : "))
modulo3 = nombre%3
modulo2 = nombre%2
if modulo2 == 0 :
    print ("Ce nombre est pair")
elif modulo3 == 0 :
    print ("Ce nombre est impair, mais est multiple de 3")
else :
    print ("Ce nombre n'est ni pair ni multiple de 3")
```

**Exercice 4. Moyenne pondérée (3 pts)**

Écrire un programme qui affiche à partir des deux listes saisies au clavier la moyenne pondérée d'un étudiant.

Exemple d'exécutions:

> \>\>\>
>
> Entrez un note (taper Q pour terminer) : 11
>
> Entrez un note (taper Q pour terminer) : 12
>
> Entrez un note (taper Q pour terminer) : 13
>
> Entrez un note (taper Q pour terminer) : Q
>
> [11.0, 12.0, 13.0]
>
> Entrez le coefficient : 1
> 
> Entrez le coefficient : 2
> 
> Entrez le coefficient : 3
>
> [1, 2, 3]
>
> 6
>
> La moyenne pondérée est de :  12.333333333333334
   
</div>



```
element=""
notes=[]
nombre=0
while element !="Q":
    element=input("Entrez un note(taper Q pour terminer) : ")
    if element!="Q":
        notes=notes+[float(element)]
        nombre=nombre+1

print(notes)

coef=[]
totalcoef=0
while nombre >0:
    c=int(input("Entrez le coefficient : "))
    coef=coef+[c]
    totalcoef = totalcoef + c
    nombre=nombre-1
    
print(coef)
print(totalcoef)

total=0
for i in range(0,len(notes)):
    total= total+notes[i]*coef[i]

mp= total/totalcoef
print("la moyenne pondérée est de : ", mp)
```

**Exercice 6. Fonction (2 pts)**

Un vigile sécurité de la fac travaille chaque semaine dans un bâtiment différent.
Il doit faire une ronde $n$ fois par jour. Le bâtiment fait $n$ étages où chaque palier compte $m$ marches de 16 cm.   Son bureau est au $n$ième étage.
    
On choisit les paramètres suivant : 
    
•	Le nombre de ronde : r
    
•	Le nombre d'étage : e
    
•	Le nombre de jour : j
    
•	Le nombre de marche : m
    
Écrire une **fonction(r,e,jm)**  qui calcule la hauteur parcourue en une semaine.  
On écrira ensuite une procédure pour saisir les paramètres nécessaires au clavier avant d'appeler cette fonction.
    
Une semaine comporte 5 ou 6 jours ouvrés (selon les semaines) et n'oublions pas qu'une fois en bas, le vigile doit remonter.


    
Exemple d'exécution:


> \>\>\>
>
> Entrer Le nombre de ronde par jour : 5
>
> Entrer Le nombre d'étage de l'immeuble : 5
>
> Entrer Le nombre de jour ouvré : 5
>
> Entrer Le nombre de marche par palier : 20
>
> Le vigile a parcouru en cm:  80000  cm
>
> Le vigile a parcouru en m:  800.0  m
>


   
</div>

```
def hauteurparcourue(r,e,j,m):
    #16 cm la hauteur d'une marche
    #2 l'aller et le retour
    h= r*e*j*m*16*2
    return h
    

ronde = int(input("Entrer Le nombre de ronde par jour : "))
etage = int(input("Entrer Le nombre d’étage de l'immeuble  : "))
jour = int(input("Entrer Le nombre de jour ouvré : "))
marche = int(input("Entrer Le nombre de marche par palier : "))


distparcourue = hauteurparcourue(ronde,etage,jour,marche)

print("Le vigile a parcouru en cm: ",distparcourue, " cm")
print("Le vigile a parcouru en m: ",distparcourue/100, " m")
```

**Exercice 7. Somme (2 pts)**

Écrire un programme qui possède une fonction **sommesuite(valeur)** qui réalise la somme de toute les valeurs de 0 à valeur.

Exemple d'exécution:

> \>\>\>
>
> La valeur de départ : 10
>
> la somme suite est égale à :  55


</div>

```
def sommesuite(borne):
    somme = 0
    while borne >=0:
        somme =somme +borne
        borne = borne -1
    return somme


valeur=int (input("La valeur de départ : "))

print ("la somme suite est égale à : ", sommesuite(valeur))
```

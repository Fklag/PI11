**Exercice 3. Structures de contrôles (2 pts)**

Écrire un programme qui affiche en fonction d'une valeur saisie l'un des messages suivants :  
- "Ce nombre est pair"  
- "Ce nombre est impair, mais est multiple de 3"
- "Ce nombre n'est ni pair ni multiple de 3"
    
Un nombre **b** est multiple de 3 si le reste de la division de ce nombre par 3 est égal à 0. Ce reste est nommé modulo.   
    
*Syntaxe Python*  
$\%$ permet d'obtenir le modulo en Python



Exemple d'exécution:

> \>\>\>
>
> Nombre : 8
>
